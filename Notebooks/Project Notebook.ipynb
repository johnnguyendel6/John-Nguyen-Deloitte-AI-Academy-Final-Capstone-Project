```python
#Import Necessary Libraries and Packages
import numpy as np 
import pandas as pd  
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

### Comparing Tech Revenue to Profit

#Read the Fortune 100 2021 CSV File
df = pd.read_csv(r"C:\Users\johnnguyen6\Documents\John-Nguyen-Deloitte-AI-Academy-Final-Capstone-Project\Data(Pre-Processed)\Data(Processed)\Consolidated 2021 Fortune 100 Company Data.csv", header=0)

#Drop Unnecessary Columns
df = df.drop([' Rank ',' Name ',' Source '], axis = 1)

#Filter Tech Data
df = df[df[" Group "] == "Yes"]

#Drop Group "Yes"
df = df.drop([' Group '], axis = 1)


#Columns to Be Analyzed
Independent_x = df[' Revenue ']
Dependent_y = df[' Profit ']


#Remove Dollar Signs, Commas, and Spaces From Strings for Revenue
Independent_x = df[' Revenue '] = df[' Revenue '].str.replace('$', '')
Independent_x = df[' Revenue '] = df[' Revenue '].str.replace(' ', '')
Independent_x = df[' Revenue '] = df[' Revenue '].str.replace(',', '')

#Remove Dollar Signs, Commas, and Spaces From Strings for Profit
Dependent_y = df[' Profit '] = df[' Profit '].str.replace('$', '')
Dependent_y = df[' Profit '] = df[' Profit '].str.replace(' ', '')
Dependent_y = df[' Profit '] = df[' Profit '].str.replace(',', '')
Dependent_y = df[' Profit '] = df[' Profit '].str.replace('(', '-')
Dependent_y = df[' Profit '] = df[' Profit '].str.replace(')', '')


#Convert Data Type to a Float
Independent_x = Independent_x.astype(float)
Dependent_y = Dependent_y.astype(float)

#Find Size Of DataFrame
sizeOfData = len(df[' Revenue '])

#Take the Average of the Panda DataFrame For Tech for both Revenue & Profit 2021
revenueMean_2021 = Independent_x.mean()
profitMean_2021 = Dependent_y.mean()

#Print Out the Average Revenue and Profit for 2021 to the User
print("The average revenue for 2021 in the Tech market is %f million dollars" %revenueMean_2021)
print("The average profit for 2021 in the Tech market is %f million dollars" %profitMean_2021)

#Read the Fortune 100 2022 CSV File
df2 = pd.read_csv(r"C:\Users\johnnguyen6\Documents\John-Nguyen-Deloitte-AI-Academy-Final-Capstone-Project\Data(Pre-Processed)\Data(Processed)\Consolidated 2022 Fortune 100 Company Data.csv", header=0)

#Drop Unnecessary Columns
df2 = df2.drop([" Rank ", " Name "," Source "], axis = 1)

#Filter Tech Data
df2 = df2[df2[' Tech '] == "Yes"]

#Drop Group "Yes"
df2 = df2.drop([' Tech '], axis = 1)

#Columns to Be Analyzed
Independent_x2 = df2[' Revenue ']
Dependent_y2 = df2[' Profit ']

#Remove Dollar Signs, Commas, and Spaces From Strings for Revenue
Independent_x2 = df2[' Revenue '] = df2[' Revenue '].str.replace('$', '')
Independent_x2 = df2[' Revenue '] = df2[' Revenue '].str.replace(' ', '')
Independent_x2 = df2[' Revenue '] = df2[' Revenue '].str.replace(',', '')

#Remove Dollar Signs, Commas, and Spaces From Strings for Profit
Dependent_y2 = df2[' Profit '] = df2[' Profit '].str.replace('$', '')
Dependent_y2 = df2[' Profit '] = df2[' Profit '].str.replace(' ', '')
Dependent_y2 = df2[' Profit '] = df2[' Profit '].str.replace(',', '')
Dependent_y2 = df2[' Profit '] = df2[' Profit '].str.replace('(', '-')
Dependent_y2 = df2[' Profit '] = df2[' Profit '].str.replace(')', '')

#Convert Data Type to a Float
Independent_x2 = Independent_x2.astype(float)
Dependent_y2 = Dependent_y2.astype(float)

#Find Size of PandaFrame
sizeOfData_2 = len(df2[' Revenue '])

#Take the Average of the Panda DataFrame For Tech for 2022
revenueMean_2022 = Independent_x2.mean()
profitMean_2022 = Dependent_y2.mean()

#Print Out the Average Revenue and Profit for 2021 to the User
print("The average revenue for 2022 in the Tech market is %f million dollars" %revenueMean_2022)
print("The average profit for 2022 in the Tech market is %f million dollars" %profitMean_2022)

#Create Array to store Revenue and Profit Values for 2021 and 2022
array_Revenue = []
array_Profit = []

# Append Revenue to said Array to store Revenue and Profit Values for 2021 and 2022
array_Revenue.append(revenueMean_2021)
array_Revenue.append(revenueMean_2022)
array_Revenue = np.array(array_Revenue).reshape((-1,1))

# Append Revenue to said Array to store Revenue and Profit Values for 2021 and 2022
array_Profit.append(profitMean_2021)
array_Profit.append(profitMean_2022)
array_Profit = np.array(array_Profit)

#Implement the Linear Regression Function from Sklearn
regressionModel = LinearRegression()

#Fit the Model given the stored values of Revenue and Profit within the array
regressionModel.fit(array_Revenue,array_Profit)

#Predict Values given Revenue for 2021 and 2022
regressionModel.predict(array_Revenue)

#Calcualte the Coeficent of Determination(R2 Percentage)
score = regressionModel.score(array_Revenue,array_Profit)
print("The Coefficent of Determination for this Regression Model is %d" %score)

#Y intercept
intercept = regressionModel.intercept_
print("The Y Intercept for this Linear Regression model is %f" %intercept)

#Slope
slope = regressionModel.coef_
print("The given slope for this Linear Regression model is %f" %slope)

#Plot a Line Graph over 500000 points in intervals of 24 given the equation of the line from the linear regression model
x = np.linspace(-1000,600000,24)
y = (slope*x+intercept)

#Get the cross validation score for the data using 2 folds using root mean squared error
cross_Score = cross_val_score(regressionModel,array_Revenue,array_Profit,cv=2,scoring='neg_root_mean_squared_error')
print("The Cross Validation Score for this model is %f meaning the prediction model is overfitting" %cross_Score[0])
print("Overfitting occurs when a statistical model or machine learning algorithm captures the noise of the data thus to reduce noise and have a more accurate model we can accomodate for the noise and have a more concise model ")
plt.scatter(Independent_x,Dependent_y, label=str(2021))
plt.scatter(Independent_x2,Dependent_y2, label=str(2022))
plt.plot(array_Revenue,regressionModel.predict(array_Revenue))
plt.plot(x,y)
plt.title("Revenue Against Profits 2021-2022 Tech(Overfitting graph)")
plt.xlabel("Revenues in Millions")
plt.ylabel("Profits in Millions")
plt.annotate('R2 score = %(score)d  ' %{"score":score}, xy = (0,155000))
plt.legend(loc = 0, numpoints = 1, ncol = 2, fontsize = 8)
plt.show()

#Fix graph resolution using the cross_val score (Shift graph by cross_val_score)
print("To accomdate for outliers in the data we subtracted the score from the y-intercept to shift the model laterally")
plt.scatter(Independent_x,Dependent_y, label=str(2021))
plt.scatter(Independent_x2,Dependent_y2, label=str(2022))
plt.plot(array_Revenue,regressionModel.predict(array_Revenue))
intercept = intercept - cross_Score[0]
y = (slope*x+intercept)
plt.plot(x,y)
plt.title("Revenue Against Profits 2021-2022 Tech(Cross-validated Graph)")
plt.xlabel("Revenues in Millions")
plt.ylabel("Profits in Millions")
plt.annotate('R2 score = %(score)d  ' %{"score":score}, xy = (0,155000))
plt.legend(loc = 0, numpoints = 1, ncol = 2, fontsize = 8)
plt.show()
```
